#### Feature Engineering

* Two of the categorical features available in the dataset have over 1000 categories present, and as such, i have to deal with them, I will be trying the frequency encoding method first
* This method involves encoding the category with its frequency

```python
col = ['sending_address', 'receiving_address']
for c in col:
    freq = df[c].value_counts()
    df[c + '_freq'] = df[c].map(freq)
df
```

```python
df.drop(['sending_address', 'receiving_address'],
       axis = 1,
       inplace = True)
df
```

* The remaining categorical features will be encoded using OneHotEncoding

```python
df.info()
```

```python
df1 = df.copy()
```

```python
y = df1.anomaly
x = df1.drop('anomaly',
            axis = 1)
```

```python
x
```

```python
num_features = x.select_dtypes(exclude = 'object').columns
cat_features = x.select_dtypes(include = 'object').columns
```

```python
cat_features
```

```python
ohe = OneHotEncoder(drop = 'first',sparse = False).set_output(transform = 'pandas')
ohetransform = ohe.fit_transform(x[cat_features])
ohetransform
```

```python
new_df = pd.concat([x,ohetransform],
              axis = 1)
```

```python
new_df
```

```python
new_df.drop(cat_features,
           axis = 1,
           inplace = True)
```

```python
new_df
```

```python
my_df = pd.concat([new_df,y],
                 axis = 1)
my_df.head()
```

```python
my_df.drop(['hour_of_day','ip_prefix'],axis = 1,inplace = True)
```

```python
x = my_df.drop('anomaly',
           axis = 1)
y = my_df.anomaly
```

```python
import pickle
pickle.dump(ohe, open('encoder.pkl','wb'))
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit(x)
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
pickle.dump(scaler,open('scal.pkl','wb'))
```

```python
X_train,X_test,y_train,y_test = train_test_split(x_scaled, y,test_size=0.2,
                                                random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
rand_model = RandomForestClassifier()
rand_model.fit(X_train,y_train)
```

```python
rand_model.score(X_test,y_test)
```

```python
model_score = {}
models = {'Rand_forest': RandomForestClassifier(),
         #'Cat_model': CatBoostClassifier(),
         #'Xg_model':XGBClassifier(),
         'Grad_model':GradientBoostingClassifier(),
         'DecisionTree': DecisionTreeClassifier()}

def fit_and_score(models,X_train,X_test,y_train,y_test):
    for name, model in models.items():
        model.fit(X_train,y_train)
        model_score[name] = model.score(X_test,y_test)
    return model_score
model_score = fit_and_score(models,X_train,X_test,y_train,y_test)
```
